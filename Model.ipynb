```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('final_tel.csv')
df.head()
```

```python
df=df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
X = df.drop('Churn', axis = 1)
X
```

```python
y = df['Churn']
y
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(max_iter=1000)
log.fit(X_train,y_train)
```

```python
y_pred_log = log.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_log))
```

```python
accuracy_score(y_test, y_pred_log)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
y_pred_prob_log = log.predict_proba(X_test)[:, 1]

fpr_log, tpr_log, thresholds = roc_curve(y_test, y_pred_prob_log)
roc_auc_log = roc_auc_score(y_test, y_pred_prob_log)
```

```python
plt.plot(fpr_log, tpr_log, label='Logistic Regression (AUC = %0.3f)' % roc_auc_log)
plt.plot([0, 1], [0, 1], 'b--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_pred_dtc = dtc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_dtc))
```

```python
accuracy_score(y_test, y_pred_dtc)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
y_pred_prob_dtc = dtc.predict_proba(X_test)[:, 1]

fpr_dtc, tpr_dtc, thresholds_dtc = roc_curve(y_test, y_pred_prob_dtc)
roc_auc_dtc = roc_auc_score(y_test, y_pred_prob_dtc)
```

```python
plt.plot(fpr_dtc, tpr_dtc, label='Decision Tree (AUC = %0.3f)' % roc_auc_dtc)
plt.plot([0, 1], [0, 1], 'b--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train,y_train)
```

```python
y_pred_rfc = rfc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred_rfc))
```

```python
accuracy_score(y_test, y_pred_rfc)
```
